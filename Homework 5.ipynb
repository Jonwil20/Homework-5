```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
ride_data = pd.read_csv("data/ride_data.csv")
city_data = pd.read_csv("data/city_data.csv")
```

```python
combined_data = pd.merge(ride_data, city_data)
combined_data.head()
```

```python
y_axis = combined_data.groupby(["city"], as_index=False).fare.mean()
y_axis
```

```python
y_axis2 =y_axis["fare"].tolist()
y_axis2
```

```python
x_axis = combined_data.groupby(["city"], as_index=False).fare.count()
x_axis
```

```python
x_axis2 = x_axis["fare"].tolist()
x_axis2
```

```python
s_axis = combined_data.groupby(["driver_count"], as_index=False).fare.count()
s_axis
```

```python
s_axis2 = s_axis["driver_count"].tolist()
s_axis2
```

```python
z_axis = combined_data.groupby(["type"], as_index=False).fare.count()
z_axis
```

```python
z_axis2 = z_axis["type"].tolist()
z_axis2
```

```python
x_limit = 100
x_axis = np.arange(0, 75, 1)
y_axis = np.arange(0, 75, 1)
```

```python
plt.scatter(x_axis2, y_axis2, marker="o", facecolors="red", edgecolors="black",
            s=s_axis2, alpha=0.55)
```

```python
plt.title("Pyber Ride Sharing Data 2016")
plt.xlabel("Total Number of Rides Per City")
plt.ylabel("Average Fare")
```

```python
plt.ylim(15,45)
plt.xlim(0,40)
```

```python
labels = ["Rural", "Suburban", "Urban"]
plt.legend(z_axis2, labels=labels, scatterpoints=1, title= "Area Type")
```

```python
plt.show()
plt.grid()
plt.savefig('pybar_scatter.png')
```
